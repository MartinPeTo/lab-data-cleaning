#### 3. Create a connection to BigQuery

```python
client = bigquery.Client()
```

#### 4. Import data from the users table from the stackoverflow public database for United Kingdom only

```python
stack_users="""
SELECT
    *
    
FROM
    `bigquery-public-data.stackoverflow.users`
    
WHERE
    location LIKE 'United Kingdom' or location LIKE 'UK'
"""

```

```python
query_job=client.query(query=stack_users)
users_df=query_job.to_dataframe()
users_df.head()
```

```python
users_df.shape
```

#### 5. Rename id column to user_id

```python
users_df=users_df.rename(columns = {'id':'user_id'})
users_df.head()
```

#### 6. Import from the stackoverflow_posts table all items which have up to 20 view counts and only select 'id', 'owner_user_id', 'score', 'view_count', 'comment_count', 'favorite_count'

```python
stack_posts="""
SELECT
    id,
    owner_user_id,
    score,
    view_count,
    comment_count,
    favorite_count
    
FROM
    `bigquery-public-data.stackoverflow.stackoverflow_posts`
    
WHERE
    view_count<=20
"""
```

```python
query_job=client.query(query=stack_posts)
posts_df=query_job.to_dataframe()
posts_df.head()
```

```python
posts_df.shape
```

#### 7. Rename id column to post_id and owner_user_id to user_id

```python
posts_df=posts_df.rename(columns = {'id':'post_id', 'owner_user_id':'user_id'})
posts_df.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: user_id, reputation,views,up_votes,down_votes
    **posts columns**: post_id, score,user_id,view_count,comment_count

```python
users_newdf = users_df[['user_id', 'reputation','views','up_votes','down_votes']].copy()
users_newdf.head()
```

```python
posts_newdf = posts_df[['post_id', 'score','user_id','view_count','comment_count']].copy()
posts_newdf.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
user_post_df=pd.merge(users_newdf,posts_newdf, on ='user_id')
user_post_df.head()
```

#### 9. How many missing values do you have in your merged dataframe? 

```python
print('The users rows are: ' + str(users_newdf.shape[0]))
print('The posts rows are: ' + str(posts_newdf.shape[0]))
print('The final rows are: ' + str(user_post_df.shape[0]))
```

#### Bonus: Identify extreme values in your merged dataframe, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder. Hint: post_id cannot have outliers!

```python
stats = user_post_df.describe().T
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
#Borramos las lineas de user_id y post_id:
stats = stats.drop(['user_id','post_id'])
stats
```

```python
outliers = pd.DataFrame(columns=user_post_df.columns)

for col in stats.index:
    iqr = stats.at[col, 'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col, '25%'] - cutoff
    upper = stats.at[col, '75%'] + cutoff
    results = user_post_df[(user_post_df[col] < lower) | (user_post_df[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
    
outliers.head()
```

```python
outliers.to_csv('outliers.csv')
```
